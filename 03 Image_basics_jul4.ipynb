```python
chess = np.zeros([8,8])
for count in range(8):
    for x in range(8):
        if (count + x) % 2 == 0:
            chess[count, x] = 1
```

```python
chess
```

```python
plt.imshow(chess, cmap = 'gray');
plt.imsave('chess.jpg', chess)
```
